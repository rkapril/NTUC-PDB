```python
#import the matplotlib library

import matplotlib.pyplot as plt
```

```python
#Line plot

years = ['2020', '2021', '2022', '2023', '2024']

population = [8000000, 120000000, 3000000, 50000000, 3600000]

plt.plot(years, population)
plt.title('Years vs Population in Mars') #This adds a title to the chart
plt.xlabel('Year') #Adding x axis name
plt.ylabel('Total Population') #Adding y axis name
plt.show()
```

```python
#Dummy data

temp = [30, 32, 33, 28.7, 35, 29, 29]

ice_creams = [100, 120, 130, 75, 125, 77, 83]

plt.scatter(temp, ice_creams)

plt.xlabel('Temperature')
plt.ylabel('Ice creams')
plt.title('Relationship Between Temperature and Ice Cream sold')
plt.show()
```

```python
#Histogram

numbers = [1,2,3,3.5,4,4.5,4,6.3,6.6,6.7,7,8,9,9.2,9.3,9.5]

plt.hist(numbers, bins=10)

plt.title('Histogram')
plt.xlabel('Numbers')
plt.ylabel('Frequency')

plt.show()
```

```python
#Bar

labels = ['Javascript', 'Java', 'Python', 'R']

usage = [62.8, 43.5, 71, 25]

plt.bar(labels,usage)

plt.title('The usage of different languages')
plt.xlabel('Languages')
plt.ylabel('Usage')
plt.show()
```

```python
plt.barh(labels,usage)

plt.title('The usage of different languages')
plt.xlabel('Languages')
plt.ylabel('Usage')
plt.show()
```
