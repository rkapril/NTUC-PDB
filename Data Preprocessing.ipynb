```python
#drop two columns: LAW_CODE, ARREST_BORO
df.drop(['LAW_CODE','ARREST_BORO'],axis=1)
df.drop(['LAW_CODE','ARREST_BORO'],axis=1).columns
```

```python
columns_to_drop = ['LAW_CODE', 'ARREST_BORO']
df = df.drop(columns_to_drop, axis='columns')

df.head()
```

```python
#create a new column in the dataframe which is called Jail time

#If Felony (F) - 36 months
#If Misfameonors (M) - 12 months
#If Violation (V) - 1 month
```

```python
#np.where

df['JAIL_TIME'] = np.where(df['LAW_CAT_CD']=="F", 36,
                          np.where(df['LAW_CAT_CD']=="M", 12,
                                  np.where(df['LAW_CAT_CD']=="V",1,0)))

df.head()
```
