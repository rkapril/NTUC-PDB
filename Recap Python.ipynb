### Flow Control

```python
#Flow Control is about how program flows 
#comprising of conditional statements, loops etc.

#You go to a grocery store to buy milk and eggs
#You buy a jar of milk ansd check if the store contains eggs
#If it contains eggs, you buy a dozen

#defining the variables

milk = 0
eggs = 0
store_has_eggs = False
milk = milk+1

#<=, >=, ==

if store_has_eggs == True:
    eggs = eggs+12
    
print('Milk is', milk)
print('Eggs are', eggs)
```

```python
#Loops

#While loop
number = 43

while number > 0:
    number = number-5
    print(number)

  
```

```python
#For loop
for i in range(1,11):
    print(i)  
```
