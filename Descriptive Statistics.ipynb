```python
#We will use dummy data and see how we can use 
#the various descriptive stats in the dataframe
```

```python
#import the library

import pandas as pd
```

```python
#Let's create a dummy dataframe

df = pd.DataFrame({'name': ['Jack', 'John', 'Joe'],
                  'salary': [1500, 25000, 125000]})

df
```

```python
#calculate the average salary

df['salary'].mean()
```

```python
#median salary

df['salary'].median()
```

```python
#Let's create antoher dataframe

newdf = pd.DataFrame({'name': ['Jack', 'John', 'Joe', 'Megan', 'Jim'],
                     'salary': [1500, 25000, 125000, 1500, 1500]})

newdf
```

```python
#calculate the mode

newdf['salary'].mode()
```

```python
#boxplot
df.boxplot(column='salary')
```
